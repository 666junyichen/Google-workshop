```python
# üß© ImagineX Mini Challenge ‚Äì Poster Creator
# Everyone can run this ‚Äî no setup, no install, no coding experience needed!

from PIL import Image, ImageDraw, ImageFont
from IPython.display import display
from google.colab import files

# üëâ Step 1: Type your idea below (one sentence)
topic = input("üí° Describe your idea in one sentence: ")

# üëâ Step 2: Choose mode
#    Type A to upload your own image OR B to auto-generate a simple drawing
MODE = input("Choose mode: A = Upload image, B = Simple drawing  --> ").strip().upper() or "B"

# --- fonts ---
def F(name="DejaVuSans.ttf", size=42, bold=False):
    try:
        if bold: name="DejaVuSans-Bold.ttf"
        return ImageFont.truetype(name, size)
    except:
        return ImageFont.load_default()

# --- colours ---
CREAM=(246,242,232); GREEN=(40,120,90)
SKY=(201,226,236); LEAF1=(141,185,149); LEAF2=(109,160,138); LEAF3=(85,137,118)
BROWN=(92,62,44); YELLW=(244,214,60); SKIN=(232,190,150); DARK=(34,34,34)

# --- make poster ---
W,H=360,562
poster=Image.new("RGB",(W,H),(28,28,28))
d=ImageDraw.Draw(poster)
d.rounded_rectangle([(8,8),(W-8,H-8)],radius=18,fill=CREAM)

if MODE=="A":
    print("üìÅ Upload any photo or artwork you want inside your poster:")
    up=files.upload()
    fname=next(iter(up.keys()))
    img=Image.open(fname).convert("RGB")
    win=(22,22,W-22,410)
    wx1,wy1,wx2,wy2=win
    ww,wh=wx2-wx1,wy2-wy1
    img=img.resize((ww,wh))
    poster.paste(img,(wx1,wy1))
else:
    # Simple default drawing
    d.rectangle([(22,22),(W-22,410)],fill=SKY)
    d.text((30,36),"YOUR",font=F(size=48,bold=True),fill=GREEN)
    d.text((30,96),"POSTER",font=F(size=48,bold=True),fill=GREEN)
    def tree(x,y,s=1.0):
        d.ellipse([(x-46*s,y-26*s),(x+46*s,y+64*s)],fill=LEAF1)
        d.ellipse([(x-80*s,y-20*s),(x-10*s,y+50*s)],fill=LEAF2)
        d.ellipse([(x+12*s,y-12*s),(x+80*s,y+56*s)],fill=LEAF3)
        d.rectangle([(x-6*s,y+44*s),(x+6*s,y+116*s)],fill=BROWN)
    tree(100,250,1.0); tree(260,255,1.0); tree(300,280,0.8)
    # stick-figure person
    d.ellipse([(140,210),(190,255)],fill=SKIN)
    d.rectangle([(118,268),(212,320)],fill=YELLW)
    d.polygon([( 96,336),(236,336),(248,392),(84,392)],fill=DARK)

# bottom labels
d.text((46,456),"Date", font=F(size=26,bold=True),fill=GREEN)
d.text((158,456),"Time", font=F(size=26,bold=True),fill=GREEN)
d.text((268,456),"Place",font=F(size=26,bold=True),fill=GREEN)
for i,x in enumerate([46,158,268]):
    d.line([(x,490),(x+56,490)],fill=GREEN,width=5)

# show + save
display(poster)
poster.save("/content/your_poster.png",quality=95)
print("\n‚úÖ Poster saved to /content/your_poster.png")
print(f"Your idea: {topic}")
```
